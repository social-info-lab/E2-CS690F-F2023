```python
import pandas as pd
import string
import re
import nltk
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
df=pd.read_csv(r"E:\UMass\Academics\Fall2023\690F\Project\Data\DataStephen.csv")
```

```python
df.info()
```

```python
def remove_punct(text):
    text  = "".join([char for char in text if char not in string.punctuation])
    text = re.sub('[0-9]+', '', text)
    return text
```

```python
df['text_punc_rmvd'] = df['text'].apply(lambda x: remove_punct(x))
df.head(10)
```

```python
def tokenization(text):
    text = re.split('\W+', text)
    return text

df['text_tokenized'] = df['text_punc_rmvd'].apply(lambda x: tokenization(x))
df.head()
```

```python
stopword = nltk.corpus.stopwords.words('english')
```

```python
def remove_stopwords(text):
    text = [word for word in text if word not in stopword]
    return text
    
df['text_nonstop'] = df['text_tokenized'].apply(lambda x: remove_stopwords(x))
df.head(10)
```

```python
wn = nltk.WordNetLemmatizer()

def lemmatizer(text):
    text = [wn.lemmatize(word) for word in text]
    return text

df['text_lemmatized'] = df['text_nonstop'].apply(lambda x: lemmatizer(x))
df.head()
```

```python
nltk.download('wordnet')
```

```python
def sentence(x):
#print(x)
    return ' '.join(x)
df['sent'] = df['text_lemmatized'].apply(lambda x: sentence(x))
df.head(5)
```

```python
vec = TfidfVectorizer()
```

```python
matrix = vec.fit_transform(df['sent'])
                           
```

```python
arr = matrix.toarray()
```

```python
features = vec.get_feature_names()
```

```python
df_tf_idf = pd.DataFrame(arr, columns = features)
```
